```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.Figure()

axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis2 = fig.add_axes([0.2,0.5,0.4,0.3])

axis1.plot(x,y,'b')
axis1.set_title('BIG PLOT')
axis1.set_xlabel('x values')
axis1.set_ylabel('y values')
axis2.plot(y,x,'r')
axis2.set_title('small plot')
axis2.set_xlabel('y values')
axis2.set_ylabel('x values')
fig
```
